```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load datasets
user_details = pd.read_csv('C:/Users/Rambabu/Desktop/UserDetails.csv')  
cooking_sessions = pd.read_csv('C:/Users/Rambabu/Desktop/CookingSessions.csv')
order_details = pd.read_csv('C:/Users/Rambabu/Desktop/OrderDetails.csv')

# Inspect datasets
print(user_details.head())
print(cooking_sessions.head())
print(order_details.head())
```

### Data Cleaning and Handle missing values

```python
# Handle missing values
user_details.fillna('Unknown', inplace=True)  # Fill missing values in user details with 'Unknown'
cooking_sessions.dropna(inplace=True)         # Drop rows with missing values in cooking sessions
order_details.ffill(inplace=True)             # Use forward fill for missing values in order details
```

### Merging Data

```python
merged_data = pd.merge(user_details, cooking_sessions, on='User ID', how='inner')
merged_data
```

### Final Dataset

```python
final_data = pd.merge(merged_data, order_details, on='Session ID', how='inner')
final_data
```

## Data VIsualisation

### Popular dishes

```python

# Popular dishes
popular_dishes = final_data['Dish Name_y'].value_counts()
popular_dishes
```

```python

# Visualizing popular dishes
plt.figure(figsize=(10, 6))
sns.barplot(x=popular_dishes.index[:10], y=popular_dishes.values[:10], palette='viridis')
plt.title('Top 10 Popular Dishes')
plt.xticks(rotation=45)
plt.xlabel('Dish Name')
plt.ylabel('Order Count')
plt.show()
```

```python
age_groups = final_data.groupby('Age')['Total Orders'].sum()
plt.pie(age_groups, labels=age_groups.index, autopct='%1.1f%%', startangle=140)
plt.title('Orders by Age Group')
plt.show()
```

#### Relationship Between Cooking Sessions and User Orders

```python
# Scatter plot for visualizing trends
plt.figure(figsize=(10, 6))
sns.scatterplot(data=final_data, x='Session ID', y='Total Orders',hue='Age', palette='Set2')
plt.title('Relationship Between Cooking Sessions and Orders')
plt.xlabel('Number of Cooking Sessions')
plt.ylabel('Number of Orders')
plt.legend(title='Age Group')
plt.show()
```

#### Trend Analysis Over Time

```python
# Regional analysis
region_orders = final_data.groupby('Location')['Total Orders'].sum().sort_values(ascending=False)

# Bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=region_orders.index, y=region_orders.values, palette='muted')
plt.title('Order Count by Region')
plt.xticks(rotation=45)
plt.xlabel('Location')
plt.ylabel('Total Orders')
plt.show()
```

### Average Orders by Age Group and Region

```python
# Group by age group and region, and calculate average order count
age_region_analysis = final_data.groupby(['Age', 'Location'])['Total Orders'].mean().unstack()

# Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(age_region_analysis, cmap='YlGnBu', annot=True, fmt=".1f")
plt.title('Average Orders by Age Group and Region')
plt.xlabel('Region')
plt.ylabel('Age Group')
plt.tight_layout()
plt.show()
```

### Cooking Preferences by age group

```python
# Popular dishes by age group
dishes_by_age = final_data.groupby(['Age', 'Dish Name_y'])['Total Orders'].sum().unstack()

# Stacked bar chart
dishes_by_age.plot(kind='bar', stacked=True, figsize=(12, 8), colormap='viridis')
plt.title('Cooking Preferences by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Order Count')
plt.legend(title='Dishes', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### User Engagement Distribution

```python
# Engagement analysis
final_data['engagement'] = np.where(final_data['Total Orders'] > 5, 'Active', 'Inactive')
engagement_counts = final_data['engagement'].value_counts()

# Pie chart
plt.figure(figsize=(8, 8))
plt.pie(engagement_counts, labels=engagement_counts.index, autopct='%1.1f%%', startangle=140, colors=['#4CAF50', '#FFC107'])
plt.title('User Engagement Distribution')
plt.show()
```

### Top 10 Users by Order Frequency

```python
# Top users
top_users = final_data.groupby('User ID_x')['Total Orders'].sum().sort_values(ascending=False).head(10)

# Horizontal bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=top_users.values, y=top_users.index, palette='coolwarm')
plt.title('Top 10 Users by Order Frequency')
plt.xlabel('Order Count')
plt.ylabel('User ID')
plt.show()
```
